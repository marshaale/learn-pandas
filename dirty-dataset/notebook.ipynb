```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("dirty-movies.csv")
df
```

```python
df['Tour title'] = df['Tour title'].replace(r'[^A-Za-z]','',regex=True)
df['Tour title'] = df['Tour title'].str.strip()
df['Tour title'] = df['Tour title'].str.title()
```

```python
df['Artist'] = df['Artist'].replace(r'[^A-Za-z]','',regex=True)
df['Artist'] = df['Artist'].str.strip()
df['Artist'] = df['Artist'].str.title()
```

```python
df
```

```python
df['Year(s)'] = df['Year(s)'].str.slice(0,4)
df['Year(s)']
```

```python
df.columns.tolist()
```

```python
df = df.rename(columns={'Year(s)':'Year','Actual\xa0gross':'Actual gross','Adjusted\xa0gross (in 2022 dollars)':'Adjusted gross'})
df
```

```python
df.columns.tolist()
```

```python
def remove_str_chars(col):
    df[col] = df[col].astype(str)
    df[col] = df[col].str.strip()
    df[col] = df[col].str.replace(r"[^0-9.]", "", regex=True)

numeric_columns = ['Actual gross','Adjusted gross', 'Average gross', 'Peak','All Time Peak']

for col in numeric_columns:
    if not col in df.columns.tolist():
        raise KeyError(f"{col} not found")
    remove_str_chars(col)
df
```

```python
def to_numeric(col):
    df[col] = pd.to_numeric(df[col],errors="coerce")

for col in numeric_columns:
    if not col in df.columns.tolist():
        raise KeyError(f"{col} not found")
    to_numeric(col)

df
```

```python
df = df.drop(columns=['Ref.'])
```

```python
df
```

```python
df[['Peak','All Time Peak']] = df[['Peak','All Time Peak']].interpolate()
df
```

```python
df.to_csv("clean_movies.csv")
```
