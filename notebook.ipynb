```python
def stander_name(col):
    df[col] = df[col].astype(str).str.strip()
    df[col] = df[col].astype(str).str.title()

stander_name('city')
df
```

```python
def numeric_fix(col):
    df[col] = pd.to_numeric(df[col],errors="coerce")

numeric_fix('temperature_c')
numeric_fix('humidity')
numeric_fix('wind_speed')
numeric_fix('rainfall_mm')
df
```

```python
df.isna().sum()
```

```python
def fill_none(col):
    df.fillna({ col: df[col].median() },inplace=True)

fill_none('humidity')
fill_none('temperature_c')
fill_none('wind_speed')
fill_none('rainfall_mm')

df
```
