# Conditional selection
- condition_mask is the where condition like in sql
- columns is like select with specified columns in sql
- in sql mind looks like  select date,city,temperature_c from df where temperature_c > 22
- what about * all columns just remove columns

```python
condition_mask = df['temperature_c'] > 22
columns = ['date','city','temperature_c']
print(df[condition_mask])
df[columns][condition_mask]
```
