# Conditional selection
- condition_mask is the where condition like in sql
- columns is like select with specified columns in sql
- in sql mind looks like  select date,city,temperature_c from df where temperature_c > 22
- what about * all columns just remove columns

```python
condition_mask = df['temperature_c'] > 22
columns = ['date','city','temperature_c']
print(df[condition_mask])
df[columns][condition_mask]
```

# Ways to create DataFrame

```python
# csv
df = pd.read_csv("weather.csv")
df
```

```python
# dictionary
weather_data = {
    "date":['2026-01-03','2025-10-23'],
    "city":["mogadishu","Kismayo"],
    "temperature":[31,28]
}

df = pd.DataFrame(weather_data)
df
```

```python
# tuple list
weather_data = {
    ('2026-01-08','mogadishu',29),
    ('2025-09-11','jowhar',28)
}

df = pd.DataFrame(weather_data,columns=['date','city','temperature'])
df
```

```python
# dictionary list
weather_data = [
    {'city':'mogadishu','temperature':22},
    {'city':'london','temperature':18},
    {'city':'paris','temperature':23}
]

df = pd.DataFrame(weather_data)
df
```
