# 2 Standard and consistence

### 1 For continent it can be a categorical
- str strip,title etc.
- type cast to category
- get unique list of continent
- give llm(Chatgpt) unique continent to get mapper

```python
def strip_and_title(col):
    df[col] = df[col].astype(str)
    df[col] = df[col].str.strip()
    df[col] = df[col].str.title()

def convert_to_category(col):
    df[col] = df[col].astype('category')

strip_and_title('continent')
convert_to_category('continent')

df['continent']
```

```python
df['continent'].unique().tolist()
```

```python
continent_mapper = {
    # Africa
    "Africa": "Africa",
    "Afri Ca": "Africa",
    "Afrika": "Africa",
    "Af": "Africa",

    # Asia
    "Asia": "Asia",
    "Asiya": "Asia",
    "Aisa": "Asia",
    "As": "Asia",

    # Europe
    "Europe": "Europe",
    "Eu Rope": "Europe",
    "Eur0Pe": "Europe",
    "Eu": "Europe",

    # North America
    "North America": "North America",
    "N. America": "North America",
    "North Am.": "North America",

    # South America
    "South America": "South America",
    "South Am.": "South America",
    "S. America": "South America",
    "Sa": "South America",

    # Oceania
    "Oceania": "Oceania",
    "Oce Ania": "Oceania",
    "Ocenea": "Oceania",
    "Oc": "Oceania",
    
}
```

```python
df['continent'] = df['continent'].map(continent_mapper)
```
